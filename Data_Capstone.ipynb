How many country have Republic in their full name.

```python
count=0
def counting(txt):
    global count # Initialize count as a global variable
    if 'republic' in txt.lower():
        count += 1
    return txt

df['coountry-long'] = df['country_long'].apply(counting)
print(count)
```
